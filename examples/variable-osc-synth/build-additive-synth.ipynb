```python
import maxpy as mp
```

```python
def get_num_additive_oscs(num_components):
    #based on input number, choose number of oscs to put in
    #lower limit of 4
    #if input number is odd, subtract 1 to make it even 
    if num_components < 4:
        return 4
    
    elif num_components %2 == 1:
        return num_components - 1
    else:
        return num_components


def get_additive_osc_freq_exprs(num_oscs):
    right_side = []
    left_side = []
    mult_order = [1., 2., 0.5]
    for i in range(int(num_oscs/2)):
        add_mult = i % 5 + 1
        
        right_mult = mult_order[i%3]
        expr_right = "$f1*" + str(right_mult) + " + $f2*" + str(add_mult)
        right_side.append(expr_right)
        
        left_mult = mult_order[(i+1)%3]
        expr_left = "$f1*" + str(left_mult) + " - $f2*" + str(add_mult)
        left_side.insert(0, expr_left)
    return left_side + right_side


def get_additive_osc_pan_exprs(num_oscs):
    right_side = []
    left_side = []
    
    for i in range(int(num_oscs/2)):
        expr_right = "64. + $f1*" + str(float(i+1)) + "/" + str(float(num_oscs))
        expr_left = "64. - $f1*" + str(float(i+1)) + "/" + str(float(num_oscs))
        right_side.append(expr_right)
        left_side.insert(0, expr_left)
    return left_side + right_side
```

```python
def build_additive_synth(num_comps, filename = "additive"):
    #get osc info
    
    #get number of oscs to put in 
    num_oscs = get_num_additive_oscs(num_comps)
    
    #get expr strings for frequency calculations  
    osc_freq_exprs = get_additive_osc_freq_exprs(num_oscs)
    
    #get expr strings for pan calculation 
    osc_pan_exprs = get_additive_osc_pan_exprs(num_oscs)

    #make osc patches
    osc_objs = []
    for i in range(num_oscs):
        
        #load in template osc file 
        osc = mp.MaxPatch(load_file="osc-to-add.maxpat", verbose=False)
        
        #find the two expr objects to be edited 
        pan_expr = osc.objs['obj-3']
        freq_expr = osc.objs['obj-9']
        
        #edit the expr objects, put in the freq and pan strings 
        pan_expr.edit(text=osc_pan_exprs[i], text_add="replace")
        freq_expr.edit(text=osc_freq_exprs[i], text_add="replace")
        
        #save edited osc file as a numbered file 
        osc_filename = filename + "-osc-" + str(i)
        osc.save(filename + "-osc-" + str(i), verbose=False, check=False)
        
        #save new osc filename to list 
        osc_objs.append(osc_filename)

        
        
    #open template additive patch
    additive = mp.MaxPatch(load_file = "additive-dynamic-template.maxpat", verbose=False)
    
    #find relevant objs in additive patch
    dial = None
    freqnum = None
    env = None
    loadbang = None
    snapshot = None
    bottom = None
    for key, val in additive.objs.items():
        if val.name == "dial":
            dial = val
        elif val.name == "flonum":
            freqnum = val
        elif val.name == "function":
            env = val
        elif val.name == "loadbang":
            loadbang = val
        elif val.name == "snapshot~":
            snapshot = val
        elif val.name == "additive-bottom":
            bottom = val

            
    #place oscs at y=700
    additive.set_position(0, 700)
    
    #for each osc, place and connect
    for osc_name in osc_objs:
        #place 
        osc = additive.place([osc_name], verbose=False)[0]
        
        #connect
        additive.connect((dial.outs[0], osc.ins[0]),    #get detune from dial 
                        (freqnum.outs[0], osc.ins[1]), #get frequency
                        (loadbang.outs[0], osc.ins[2]), #get loadbang 
                        (env.outs[1], osc.ins[3]),      #get env trigger
                        (snapshot.outs[0], osc.ins[4]), #get pan lfo signal
                        (osc.outs[0], bottom.ins[0]),   #send to bottom1
                        (osc.outs[1], bottom.ins[1]), verbose=False) #send to bottom2

    #save additive patch
    additive.save(filename, verbose=True, check=False)
```

```python
num_comps = 20
build_additive_synth(num_comps, filename="synth20")
```
